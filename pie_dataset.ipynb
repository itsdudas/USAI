```python
import matplotlib.pyplot as plt
import numpy as np

```

```python


# Dados
data = [669, 740, 769, 1054, 750]
colors = ['#2D2D2A', '#69995D', '#FA9F42', '#C14953', '#3AA8AC']
recipe = ["Colagem", "Defeito US", "Erro na Leitura US", "Profundidade", "Recobrimento"]

#Percentagem
total = sum(data)
percentage = [(x/total)*100 for x in data]
labels = [f"{d} ({round(p, 1)}%)" for d, p in zip(data, percentage)]

#Grfico
plt.figure(figsize=(6, 6))
plt.pie(data, labels=labels, colors=colors, wedgeprops=dict(width=0.6), startangle=0)
plt.axis('equal')
plt.legend(recipe, loc='right', bbox_to_anchor=(1.4, 0.5), shadow=False, ncol=1)
plt.show()
```
